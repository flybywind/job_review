```python
import requests
```

```python
url = "https://gist.githubusercontent.com/rob-brown/e261dde3a78f4b5e3673d65f55ff6ff5/raw/f54a619bcb4161e6256227f77babb4b18a19f6c9/access.log"
resp = requests.get(url)
resp
```

```python
lines = resp.content.decode('utf8').split('\n')
lines
```

```python
from collections import namedtuple, defaultdict

class VideoStat:
    def __init__(self, hit_count = 0, miss_count = 0):
        self.hit_count = hit_count
        self.miss_count = miss_count
    def __repr__(self):
        return f"(hc = {self.hit_count}, mc = {self.miss_count})"
        
video_dict = defaultdict(VideoStat)
```

```python
vid = video_dict['abc']
vid.hit_count += 1
vid
```

```python
ACTION_HIT = "TCP_HIT"
ACTION_MISS = "TCP_MISS"
for l in lines:
    seg = l.split()
    if len(seg) > 9:
        action, video_id = seg[6].split("/")[0], seg[9].split("/")[-2]
        vid = video_dict[video_id]
        # inside mapPartition : 
        # extract vid, do agg first 
        # output: (vid, video_stat) ==> reduce
        # inside reduce, sort inside reduce by vid
        if action == ACTION_HIT:
            vid.hit_count += 1
        if action == ACTION_MISS:
            vid.miss_count += 1
        video_dict[video_id] = vid

```

```python
sorted([t for t in video_dict.items()], key=lambda t: -t[1].miss_count)
```
